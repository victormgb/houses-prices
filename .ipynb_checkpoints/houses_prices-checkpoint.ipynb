# Dealing with null values in numerical features

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy="median")
imputer.fit(df_train_num)
df_train_num = pd.DataFrame(imputer.transform(df_train_num), columns=num_cols)

df_train_num.shape
```

```python
df_cleaned_2 = pd.concat([df_train_categorical, df_train_num, full['type_dt'].reset_index().drop(columns=['index'],axis=1)], axis=1)
df_cleaned_2.shape
```

```python
df_cleaned_2.isnull().sum().sort_values(ascending=False)
```

## Aplying one hot encoding

```python
df_cleaned = pd.get_dummies(df_cleaned_2,columns=categorical_columns)
```

```python
df_cleaned = df_cleaned.fillna(df_cleaned.mean())
```

```python
df_cleaned.isnull().sum().sort_values(ascending=False)
```

# Dividing training set and test data set

```python
X = df_cleaned.loc[df_cleaned['type_dt'] == 'train']
X_test = df_cleaned.loc[df_cleaned['type_dt'] == 'test']
X.drop(columns=['type_dt'],axis=1,inplace=True)
X_test.drop(columns=['type_dt'],axis=1,inplace=True)
```

```python
y = np.log1p(y)
plt.hist(y)
```

```python
df_cleaned.shape
```

# Select ML Algorithms to Use

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, LassoCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
estimators = {
   ' Linear': LinearRegression(),
    'Ridge': Ridge(alpha=0.5),
    'RandomForest': RandomForestRegressor(n_estimators = 10, random_state = 42),
    'GradientBoost': GradientBoostingRegressor(max_depth=8, min_samples_leaf=50, min_samples_split=500, n_estimators=360,random_state=10, subsample=0.8)
}
```

```python
best_score = 0
best_model = ''
#y_test = np.expm1(y_test)
for key in estimators:
    results = cross_validate(estimators[key],X,y,scoring='r2',cv=5,n_jobs=1)
    mean_results_score = results['test_score'].mean()
    if(mean_results_score > best_score):
        best_score = mean_results_score
        best_model = key
```

```python
best_model
```

```python
selected_estimator = estimators[best_model]
selected_estimator.fit(X,y)
selected_estimator.predict(X_test)
```

```python
y_pred = np.expm1(selected_estimator.predict(X_test))
y_test = pd.read_csv('./sample_submission.csv')['SalePrice']
```

```python
from sklearn.metrics import mean_squared_error

rms_rf = mean_squared_error(y_test.to_numpy(), y_pred, squared=False)
```

```python
rms_rf
```

```python
# Calculate mean absolute percentage error (MAPE)
mape = 100 * (rms_rf / y_test)

# Calculate and display accuracy
accuracy = 100 - np.mean(mape)

print('Accuracy:', round(accuracy, 2), '%.')
```

```python
submission = pd.DataFrame(pd.read_csv('./test.csv')['Id'])
submission['SalePrice'] = list(map(float, y_pred))
```
